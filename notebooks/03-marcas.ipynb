```python
import pandas as pd
```

```python
def _filter_brand_line(df, h):
    df_tmp = df.copy()
    df_tmp['brand-line'] = df_tmp['vehicle_make'] + df_tmp['vehicle_line'] + df_tmp['version']
    freq  = df_tmp['brand-line'].value_counts()
    lista = freq[freq>h].index.tolist()
    df_tmp = df_tmp[df_tmp['brand-line'].isin(lista)]
    return df_tmp
```

```python
data = pd.read_csv('../data/raw/car_raw.csv', index_col=False)
data= _filter_brand_line(data, 3)
```

```python
data
```

```python
dict_make = {
    make: {
        line: list(set(data[(data['vehicle_make'] == make) & (data['vehicle_line'] == line)]['version'].unique()))
        for line in set(data[data['vehicle_make'] == make]['vehicle_line'])
    }
    for make in set(data['vehicle_make'])
}
```

```python
import json
```

```python
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(dict_make, f, ensure_ascii=False, indent=4)
```
