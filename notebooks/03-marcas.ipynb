```python
import pandas as pd
```

```python
def _filter_brand_line(df, h):
    df_tmp = df.copy()
    df_tmp['brand-line'] = df_tmp['vehicle_make'] + df_tmp['vehicle_line']
    freq  = df_tmp['brand-line'].value_counts()
    lista = freq[freq>10].index.tolist()
    df_tmp = df_tmp[df_tmp['brand-line'].isin(lista)]
    return df_tmp
```

```python
data = pd.read_csv('../data/raw/car_raw.csv', index_col=False)
data= _filter_brand_line(data, 10)
```
